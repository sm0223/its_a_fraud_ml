```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=pd.read_csv("../input/train-mod/train_mod.csv")
y=pd.read_csv("../input/train-target/train_target.csv")
test=pd.read_csv("../input/test-mod/test_preprocessed.csv")
```

```python
pd.set_option('display.max_columns',None)
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus=RandomUnderSampler(random_state=0)
X_sm,y_sm=rus.fit_resample(x,y)
```

```python
params={"penalty": ["l2"],
       "C" : [0.1,0.5,1,5,10],
       "class_weight" : ["balanced", "None"],
       "max_iter" :  [10000,20000],
       "n_jobs" : [2],
       "solver" : ["saga", "lbfgs"]
       }
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score



grid_search = RandomizedSearchCV(LogisticRegression(), params, scoring = 'f1_micro', random_state=10)
grid_result = grid_search.fit(X_sm,y_sm)
```

```python
print(grid_result.best_params_)
```

```python
log = LogisticRegression(solver= 'saga', penalty= 'l2', n_jobs= -1, max_iter= 10000, class_weight= None, C= 0.1)
np.mean(cross_val_score(log,X_sm,y_sm,scoring="f1_micro", cv =5))
```

```python
ypred = log.fit(X_sm, y_sm)
```

```python
ypred = log.predict(test)
```

```python
pd.DataFrame(ypred).to_csv("log_ht_pred.csv")
```
